```python
import requests
import bs4
```

```python
request1 = requests.get('https://www.flipkart.com/lg-2025-model-ai-convertible-6-in-1-split-1-5-ton-5-star-dual-inverter-faster-cooling-energy-saving-4-way-air-swing-viraat-mode-diet-plus-ac-white/product-reviews/itm660279897729b?pid=ACNH7WZA6GH2NZJA&lid=LSTACNH7WZA6GH2NZJASKCVFB&marketplace=FLIPKART')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

### COMMENTS/REVIEWS

```python
reviews = soup.find_all('div', {'class': 'ZmyHeo'})
for review in reviews:
    print(review.get_text() + "\n\n")
```

### RATINGS

```python
ratings = soup.find_all('div', {'class': 'XQDdHH Ga3i8K'})
for rating in ratings:
    print(rating.get_text() + "  ")
```

```python
overall_rating = soup.find('div', {'class': 'ipqd2A'}).get_text()
print(overall_rating)
```

### COMMENT TAGS

```python
tags = soup.find_all('p', {'class': 'MztJPv'})
for tag in tags:
    print(tag.get_text() + "  ")
```

### CUSTOMER NAMES

```python
customers = soup.find_all('p', {'class': '_2NsDsF AwS1CA'})
for customer in customers:
    print(customer.get_text() + "  ")
```
