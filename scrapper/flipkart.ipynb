```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.findAll('div',{'class':'ZmyHeo'})
all_reviews =[]
for review in reviews:
    all_reviews.append(review.getText().replace('READ MORE','').strip());
all_reviews
```

```python
ratings = soup.findAll('div',{'class':'XQDdHH Ga3i8K'})
all_ratings = []
for rating in ratings:
    all_ratings.append(rating.getText())
all_ratings
```

```python
tags = soup.findAll('p',{'class':'z9E0IG'})
all_tags = []
for tag in tags:
    all_tags.append(tag.getText())
all_tags
```

```python
customers= soup.findAll('p',{'class':'_2NsDsF AwS1CA'})
all_customers = []
for customer in customers:
    all_customers.append(customer.getText())
all_customers
```

```python
data = {
    'names':all_customers,
    'ratings': all_ratings,
    'tags': all_tags,
    'reviews':all_reviews
}
data
```

```python
import pandas as pd
pd.set_option('display.max_colwidth',None)
df = pd.DataFrame(data)
df
```
